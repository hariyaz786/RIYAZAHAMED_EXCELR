Recommendation System

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("anime.csv")   
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
# Replace 'Unknown'
df['rating'] = df['rating'].replace('Unknown', np.nan)
df['episodes'] = df['episodes'].replace('Unknown', np.nan)
df['members'] = df['members'].replace('Unknown', np.nan)
```

```python
# Convert to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
df['members'] = pd.to_numeric(df['members'], errors='coerce')
```

```python
# Fill missing values
df['rating'].fillna(df['rating'].mean(), inplace=True)
df['episodes'].fillna(df['episodes'].median(), inplace=True)
df['members'].fillna(df['members'].median(), inplace=True)
```

```python
# Drop rows where genre is missing
df.dropna(subset=['genre'], inplace=True)
print("Data Cleaned")
```

```python
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(", "))
genre_matrix = vectorizer.fit_transform(df['genre'])
```

```python
scaler = MinMaxScaler()
num_features = scaler.fit_transform(df[['rating','episodes','members']])
```

```python
features = np.hstack((genre_matrix.toarray(), num_features))

print("Feature shape:", features.shape)
```

```python
similarity_matrix = cosine_similarity(features)

print("Similarity matrix created ")
```

```python
def recommend_anime(title, threshold=0.5, top_n=10):
    
    # Check if anime exists
    if title not in df['name'].values:
        print("Anime not found ")
        return
    
    idx = df[df['name'] == title].index[0]
    
    similarity_scores = list(enumerate(similarity_matrix[idx]))
    
    # Sort by similarity
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    # Filter based on threshold
    filtered = [i for i in similarity_scores if i[1] > threshold]
    
    # Remove 
    filtered = filtered[1:top_n+1]
    
    print(f"\nRecommendations for {title}:\n")
    
    for i in filtered:
        print(df.iloc[i[0]]['name'], " (Score:", round(i[1],2), ")")
```

```python
recommend_anime("Naruto", threshold=0.4, top_n=5)
```

```python
recommend_anime("Naruto", threshold=0.3)
recommend_anime("Naruto", threshold=0.6)
```

Interview Questions

Q1: Difference between User-Based and Item-Based Collaborative Filtering

Answer:

 Collaborative Filtering is a recommendation technique that suggests items based on user behavior and preferences.

- User-Based Collaborative Filtering :

- 
  It finds similar users based on their preferences.

  
  Then it recommends items that those similar users liked.

- Example:

- 
  If User A and User B have similar tastes,

  
  then items liked by User B will be recommended to User A.

 - Item-Based Collaborative Filtering

 - 
  It finds similar items instead of users.


  Then it recommends items similar to what the user already likes.

 Example:
 
 If a user likes Naruto,
 
 the system may recommend similar anime like Bleach or One Piece.

| User-Based                | Item-Based                  |
| ------------------------- | --------------------------- |
| Based on similar users    | Based on similar items      |
| Slower for large datasets | Faster and scalable         |
| Dynamic recommendations   | More stable recommendations |


Q2: What is Collaborative Filtering?

Answer:

Collaborative Filtering is a technique used in recommendation systems that makes predictions based on the past behavior and preferences of users.

It works on the idea that:

“Users who have similar preferences in the past will likely prefer similar items in the future.”

- Types:
  User-Based Collaborative Filtering
  Item-Based Collaborative Filtering

Example:

If:

User A likes: Naruto, Dragon Ball
    
User B likes: Naruto, Dragon Ball, One Piece
